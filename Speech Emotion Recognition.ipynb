```python
#librosa is a library that allows user to analyze music and audio.
import librosa

#soundfile is a library that allows user to read sound files.
import soundfile 

#os is a library that allows user to navigate os directory.
import os 

#glob is a library that allows user to return all file paths that match a specific pattern.
import glob 

#pickle is a library that allows user to serialize and deserialize a Python object structure.
#Seralize is the process of converting an object into a stream of bytes to store 
#the object or transmit it to memory, a database, or a file.
import pickle 

#numpy is a library that allows user to work with arrays.
import numpy as np 

#sklearn is a library for machine learning in Python.
#It contains a lot of efficient tools for machine learning and statistical modeling.
#train_test_split is a function in Sklearn model selection
#for splitting data arrays into two subsets: for training data and for testing data.
from sklearn.model_selection import train_test_split

#MLPClassifier stands for Multi-layer Perceptron classifier connects to a neural network.
from sklearn.neural_network import MLPClassifier

#accuracy_score is a function that allows users to calculate the accuracy classification score.
from sklearn.metrics import accuracy_score
```

```python
#param file_name: name of sound file
#param mfcc: Mel-frequency cepstral coefficients, the short-term power spectrum of a sound
#param chroma: the 12 different pitch classes
#param mel: Mel Spectrogram Frequency, a visualization of the frequency spectrum of a signal, 
#Purpose of function is to extract the mfcc, chroma, and mel properties of every sound file
def extract_feature(file_name, mfcc, chroma, mel):
    with soundfile.SoundFile(file_name) as sound_file:
        X = sound_file.read(dtype="float32") #convert all data to float 32 data types
        sample_rate=sound_file.samplerate #the sample rate of the sound file
        if chroma:
            stft=np.abs(librosa.stft(X)) #stft represents a signal in the time-frequency domain and returns a matrix
            #doing abs.(librosa.stft(X)) allows us to calculate the magnite of the the matrix
        result=np.array([])
        if mfcc:
            mfccs=np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0) #calculates the mean of all mfcc in file
            result=np.hstack((result, mfccs)) #hstack() is used to stack arrays in sequence horizontally (column wise)
        if chroma:
            chroma=np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0) 
            #computes a chromagram from a waveform or power spectrogram (array of chromas), and takes the mean
            result=np.hstack((result, chroma))
        if mel:
            mel=np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
            #computes a mel-spectrogram (array of magnitudes) and takes the mean
            result=np.hstack((result, mel))
    return result
```

```python
#define 8 emotions
emotions={
  '01':'neutral',
  '02':'calm',
  '03':'happy',
  '04':'sad',
  '05':'angry',
  '06':'fearful',
  '07':'disgust',
  '08':'surprised'
}
#Emotions we want our trained model to observe
observed_emotions=['calm', 'happy', 'fearful', 'disgust']
```

```python
#Identifies files that have the emotions calm, happy, fearful, and disgust.
def load_data(test_size=0.5):
    x,y=[],[]
    for file in glob.glob("C:\\Users\\SeanS\\Documents\\Speech Emotion Recognition\\speech-emotion-recognition-ravdess-data\\Actor_*\\*.wav"):
        file_name=os.path.basename(file)
        emotion=emotions[file_name.split("-")[2]]
        #if emotion not in observed_emotions:
            #continue
        feature=extract_feature(file, mfcc=True, chroma=True, mel=True)
        x.append(feature)
        y.append(emotion)
    return train_test_split(np.array(x), y, test_size=test_size, random_state=9)
    #randomly choose 1/2 of the data to train the model and the other half to test the model
```

```python
x_train,x_test,y_train,y_test=load_data(test_size=0.5)
```

```python
print((x_train.shape[0], x_test.shape[0]))
```

```python
print(f'Features extracted: {x_train.shape[1]}')
```

```python
model=MLPClassifier(alpha=0.01, batch_size=256, epsilon=1e-08, hidden_layer_sizes=(300,), learning_rate='adaptive', max_iter=500)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#Calculate the accuracy of our model
accuracy=accuracy_score(y_true=y_test, y_pred=y_pred)
#Print the accuracy
print("Accuracy: {:.2f}%".format(accuracy*100))
```
